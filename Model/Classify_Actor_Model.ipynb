### Hàm cắt khuôn mặt khi nhìn thấy đầy đủ 2 mắt

```python
def get_cropped_image_if_2_eyes(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

## Tiền xử lý:  Dùng wavelet transform để mã hóa ảnh đã cắt

```python
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    # Chuyển đổi kiểu dữ liệu
    # Chuyển đổi sang màu xám
    imArray = cv2.cvtColor(imArray , cv2.COLOR_RGB2GRAY )
    # Đổi sang kiểu float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # Tính hệ số 
    coeffs=pywt.wavedec2(imArray , mode, level=level)

    # Xử lý hệ số
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

 Có thể thấy ở trên một hình ảnh được biến đổi wavelet cung cấp các đặc điểm trên khuôn mặt như mắt, mũi, môi,.... Điều này cùng với hình ảnh pixel thô có thể được sử dụng làm đầu vào cho bộ phân loại 

### Tạo 1 dictionary chứa key là tên diễn viên và value là đường dẫn đến các ảnh diễn viên đã cắt

```python
file_names_dict = {}
for img_dir in cropped_image_dirs:
    name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(img_dir):
        file_list.append(entry.path)
    file_names_dict[name] = file_list
file_names_dict
```

### Gán cho mỗi key một ID

```python
class_dict = {}
count = 0
for name in file_names_dict.keys():
    class_dict[name] = count
    count = count + 1
class_dict
```

### Sử dụng những hình ảnh thô đã cắt cùng với những hình ảnh được biến đổi bằng wavelet để train bộ phân loại

```python
X, y = [], []
for name, training_files in file_names_dict.items():
    for training_image in training_files:
        #đọc từng file ảnh đã cắt
        img = cv2.imread(training_image)
        #resize thành 32x32
        scalled_raw_img = cv2.resize(img, (32, 32))
        #wavelet transform
        img_har = w2d(img,'db1',5)
        #resize ảnh đã watelet transform thành 32x32
        scalled_img_har = cv2.resize(img_har, (32, 32))
        #Kết hợp 2 ảnh đó thành 1 ảnh theo chiều dọc và chuyển thành dạng mảng
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        # X[i] chứa ảnh đã mã hóa, y[i] chứa id của ảnh đó.
        X.append(combined_img)
        y.append(class_dict[name]) 
```

### Dùng GridsearchCV để tìm siêu tham số.

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [1,10,100,1000], 
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,5,10]
        }
    }
}
```

```python
scores = []
best_estimators = {}
import pandas as pd
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp['model'])
    clf =  GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model': algo,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
